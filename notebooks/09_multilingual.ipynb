## 1. Offline Fallback Translations

```python
# Works without API key — uses MOCK_FALLBACK
slogan  = 'Engineering the future, one line at a time.'
company = 'NovaTech'

# Temporarily disable Gemini to test fallback
os.environ.pop('GEMINI_API_KEY', None)

results = translate_slogan(slogan, company)
results = validate_translations(results)

print(f'Original: "{slogan}"')
print()
for lang, data in results.items():
    print(f"{data['flag']} {lang:12s} ({data['native']:10s}): {data['text']}")
    print(f"   Source: {data['source']} | Tone note: {data.get('tone_note','')[:60]}…")
    print()
```

## 2. Test All 7 Core Languages

```python
test_slogans = [
    ('Think Different',              'Apple'),
    ('Just Do It',                   'Nike'),
    ('Excellence, redefined.',       'BrandSphere'),
    ('Grow what matters.',           'GreenSeed'),
    ('Your health. Our mission.',    'HealthPlus'),
]

for slogan_text, comp in test_slogans[:3]:
    r = translate_slogan(slogan_text, comp, ['Hindi','Spanish','French','German','Gujarati'])
    print(f'"{slogan_text}" ({comp})')
    for lang, d in r.items():
        print(f"  {d['flag']} {lang:10s}: {d['text']}")
    print()
```

## 3. Language Coverage Chart

```python
langs    = list(LANG_META.keys())
flags    = [LANG_META[l]['flag'] for l in langs]
# Approximate native speakers in millions (Wikipedia, 2024)
speakers = [600, 485, 280, 100, 60, 330, 920, 125, 230, 68][:len(langs)]

fig, ax = plt.subplots(figsize=(12, 5))
fig.patch.set_facecolor('#0C0D0F')
bars = ax.bar(langs, speakers, color='#C9A84C', alpha=0.85)

for bar, flag, spk in zip(bars, flags, speakers):
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+8, 
            f'{flag}\n{spk}M', ha='center', va='bottom', color='white', fontsize=9)

ax.set_ylabel('Native Speakers (millions)', color='white')
ax.set_title('BrandSphere AI — Supported Language Coverage', color='#C9A84C', fontsize=13)
ax.tick_params(axis='x', rotation=25)
total = sum(speakers)
ax.text(0.98, 0.95, f'Total reach: {total/1000:.1f}B+ speakers', 
        transform=ax.transAxes, ha='right', color='#3ECFB2', fontsize=10)
plt.tight_layout()
plt.savefig('assets/sample_exports/09_languages.png', dpi=120, bbox_inches='tight', facecolor='#0C0D0F')
plt.show()
print(f'✅ Coverage: {len(langs)} languages | {total}M+ native speakers')
```
