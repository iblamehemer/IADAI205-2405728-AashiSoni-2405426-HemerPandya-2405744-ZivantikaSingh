## 1. Build Test Assets

```python
company = 'NovaTech'
tagline = 'Engineering the future, one line at a time.'
palette = generate_palette('Technology / Software', 'Minimalist')
logos   = generate_all_logos(company, palette)
svg     = logos[0]['svg']
colors  = {name: data['hex'] for name, data in palette.items()}
print(f'Company:  {company}')
print(f'Tagline:  {tagline}')
print(f'Primary:  {colors["Primary"]}')
print(f'SVG len:  {len(svg)} chars')
```

## 2. Generate All 3 Animation Styles

```python
styles = ['typewriter', 'fade', 'slide']
gifs   = {}
for style in styles:
    gif = create_brand_gif(svg, tagline, palette, company, style, frames=20)
    gifs[style] = gif
    valid = gif[:3] == b'GIF'
    print(f'  {style:12s}: {len(gif)//1024} KB | Valid GIF: {valid}')

# Save all to disk
for style, gif_bytes in gifs.items():
    path = f'assets/sample_exports/08_anim_{style}.gif'
    with open(path, 'wb') as f:
        f.write(gif_bytes)
    print(f'  ✅ Saved {path}')
```

## 3. Frame Analysis

```python
for style, gif_bytes in gifs.items():
    img = Image.open(_io.BytesIO(gif_bytes))
    frame_count = 0
    try:
        while True:
            frame_count += 1
            img.seek(frame_count)
    except EOFError:
        pass
    
    print(f'{style:12s}: {frame_count} frames | {img.size[0]}x{img.size[1]}px | {len(gif_bytes)//1024} KB | {len(gif_bytes)//frame_count//1024} KB/frame')
```

## 5. Cross-Brand Animation Test

```python
cross_brands = [
    ('LuxeMode',  'Fashion / Apparel',        'Luxury',       'Style is a language. Speak it fluently.'),
    ('GreenSeed', 'Sustainability / Green Tech','Vibrant',      'Grow what matters.'),
    ('HealthAI',  'Healthcare',                'Professional', 'Precision care, every time.'),
]
for company_name, industry, pers, tag in cross_brands:
    pal  = generate_palette(industry, pers)
    svg_ = generate_all_logos(company_name, pal)[0]['svg']
    gif  = create_brand_gif(svg_, tag, pal, company_name, 'typewriter', frames=15)
    print(f'{company_name:12s} ({pers:12s}): {len(gif)//1024} KB GIF — {gif[:3]==b"GIF"}')
    display(IPImage(data=gif))
```
